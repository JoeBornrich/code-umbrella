 # Create a Pandas series from a list of values (\"[]\") and plot it:\n",


```python
import pandas as pd
from matplotlib import pyplot as plt

pd.Series([60, 50, 40, 30, 20]).plot(kind="bar")
```

 # Create a data frame with one column, \"ages\"\n",

```python

plotdata = pd.DataFrame({"ages": [60, 50, 40, 30, 20]})
plotdata.plot(kind="bar")
```

```python
plotdata = pd.DataFrame(
{"ages": [60, 50, 40, 30, 20]},
index=["Ada", "Bola", "Caro", "David", "Emily"])
plotdata.plot(kind="bar")
```

```python
from matplotlib import pyplot as plt
plotdata = pd.DataFrame(
{"ages": [60, 50, 40, 30, 20]},
index=["Ada", "Bola", "Caro", "David", "Emily"])
plotdata.plot(kind="bar", color = 'green')

plt.title("Sample Age of 5 Students")
plt.xlabel('Students')
plt.ylabel("Age")
```

 # Create a DataFrame with 3 columns:\n",

```python
from matplotlib import pyplot as plt
plotdata = pd.DataFrame({
"2021":[40, 12, 10, 26, 36],
"2022":[19, 8, 30, 21, 38],
"2023":[10, 10, 42, 17, 37]
      }, 
index=["Ada", "Bola", "Caro", "David", "Emily"])
plotdata.head()

 #Plot the (Stacked) Bar Chart\n",
plotdata.plot(kind="bar",stacked=False),
 
plt.title("Sample Age of 5 Students"),
plt.xlabel("Students"),
plt.ylabel("Age")  
```

# Stacked bar chart

```python
from matplotlib import pyplot as plt
plotdata = pd.DataFrame({
"2021":[40, 12, 10, 26, 36],
"2022":[19, 8, 30, 21, 38],
"2023":[10, 10, 42, 17, 37]
      }, 
index=["Ada", "Bola", "Caro", "David", "Emily"])
plotdata.head()

 #Plot the (Stacked) Bar Chart\n",
plotdata.plot(kind="bar",stacked=True),
 
plt.title("Sample Age of 5 Students"),
plt.xlabel("Students"),
plt.ylabel("Age")  
```

# DataFrame of each sales rep\n",

```python
import pandas as pd
import matplotlib.pyplot as plt

dataframe = pd.DataFrame({
    'Name': ['Awesome']*5 + ['Mensah']*5 + ['Joyce']*5,
    'Sales($)': [12, 9, 17, 19, 20, 11, 15, 12, 9, 4, 22, 19, 17, 19, 18]
})

dataframe.groupby('Name').sum().plot(kind='pie', y='Sales($)', autopct='%1.1f%%')
plt.ylabel('')  # Optional: cleans up y-axis label
plt.title('Sales Distribution by Name')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

dataframe = pd.DataFrame({
    'Name': ['Awesome']*5 + ['Mensah']*5 + ['Joyce']*5,
    'Sales($)': [12, 9, 17, 19, 20, 11, 15, 12, 9, 4, 22, 19, 17, 19, 18]
})

dataframe.groupby('Name').sum().plot(kind='pie', y='Sales($)', autopct='%1.0f%%')
plt.ylabel('')  # Optional: cleans up y-axis label
plt.title('Sales Distribution by Name')
plt.show()
```
