```python
import pandas as pd

df =pd.read_excel('Results.xlsx')
df
```

```python
del df['Total(100)']
```

```python
del df["Grade"]
```

```python
del df["Comment"]
```

```python
df
```

```python
df['Total (100)'] = df['Exam(65)'] +  df["Midterm(20)"] +  df["C.A(15)"]
```

```python
df.columns
```

```python
df['Grade'] = df["Total (100)"].apply(
    lambda x: 'A' if x >= 80 else
              'B' if x >= 70 else
              'C' if x >= 60 else
              'D' if x >= 55 else
              'E' if x >= 50 else
              'F'
)
```

```python
df
```

```python
df['Comments'] = df["Grade"].apply(
    lambda x: 'Excellent' if x == 'A' else
              'Very Good' if x =='B' else
              'Good' if x == 'C' else
              'High Average' if x == 'D' else
              'Average' if x == 'E' else
              'Fail'
)
```

```python
df
```

```python
df
```

```python
dfdecimals = 2 
df['Exam(65)'] = df['Exam(65)']. apply(lambda x : round(x, dfdecimals))
```

```python
df
```

```python
df = df.round(decimals = 1)
```

```python
df
```

```python
df.to_excel('Results_Analyzed_output.xlsx', index = False)
```
