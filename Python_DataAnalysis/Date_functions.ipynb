```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

# Convert 'mon2' to ordered categorical
df['mon2'] = pd.Categorical(df['mon2'], categories=month_order, ordered=True)

# Create the pivot
f_pivot = df.pivot_table(index='mon2',
                         values=['Profit', 'Revenues'],
                         aggfunc='sum',
                        observed = False)

# Sort by calendar month
f_pivot = f_pivot.sort_index()
f_pivot
```
