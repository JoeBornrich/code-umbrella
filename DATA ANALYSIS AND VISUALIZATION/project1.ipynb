```python
occupation_orders = df.groupby('Occupation')['Orders'].sum()
top_orders = occupation_orders.sort_values(ascending = False).head(1)
top_orders
```

```python

category_totals = df.groupby('Product_Category')['Amount'].sum()


top_10 = category_totals.sort_values(ascending=False).head(10)

total_amount = category_totals.sum()
top_10_percentage = (top_10 / total_amount) * 100

result = pd.DataFrame({
    'Total_Amount': top_10,
    'Percentage_of_Total': top_10_percentage.round(2)
})

result
```

```python
import matplotlib.pyplot as plt

result.plot(kind='bar', y='Percentage_of_Total', color = 'red',  legend=False)
plt.title('Top 10 Product Categories by % of Total Amount')
plt.ylabel('% of Total')
plt.xticks(rotation=45)
plt.tight_layout()
plt
```

```python

category_totals = df.groupby('Product_Category')['Amount'].sum()


top_10 = category_totals.sort_values(ascending=False).tail(10)

total_amount = category_totals.sum()
top_10_percentage = (top_10 / total_amount) * 100

result = pd.DataFrame({
    'Total_Amount': top_10,
    'Percentage_of_Total': top_10_percentage.round(2)
})

result
```
